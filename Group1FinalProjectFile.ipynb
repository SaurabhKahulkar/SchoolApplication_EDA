# `SchoolApplication` Class

A class representing a school application.

### Parameters:
- `date` (str): The application date.
- `name` (str): The applicant's name.
- `school` (str): The name of the school.
- `type` (str): The type of school.
- `rating` (float): The rating of the applicant.
- `tuition_fee` (float): The tuition fee for the school.
- `acceptance_rate` (float): The acceptance rate for the school.

```python
class SchoolApplication:
    def __init__(self, date, name, school, type, rating, tuition_fee, acceptance_rate):
        self.date = date
        self.name= name
        self.school= school
        self.type= type
        self.rating= rating
        self.tuition_fee= tuition_fee
        self.acceptance_rate= acceptance_rate
        self.unique_id=id(self)
        
    def __str__(self):
        return f'{self.unique_id},{self.date},{self.name},{self.school},{self.type},{self.rating},{self.tuition_fee},{self.acceptance_rate}'
```

## Loading Objects from a Pickle File

The following code snippet demonstrates loading objects from a pickle file using Python's `pickle` module.


```python
import pickle
objects = []
with (open("SchoolApplicationpkl50401.dat", "rb")) as openfile:
    while True:
        try:
            objects.append(pickle.load(openfile))
        except EOFError:
            break
```

## Writing Data to CSV

The following code writes data to a CSV file named 'data.csv'. It assumes that `objects[0]` is a list of objects with attributes that can be written to the CSV file.


```python
with open('data.csv', 'w') as f:
    f.write("uniqueId,date,name,school,type,rating,tuition_fee,acceptance_rate\n")
    for obj in objects[0]:
         f.write(str(obj)+'\n')
```

## Handling Missing Tuition Fees in DataFrame (`df`)

The data preprocessing phase involved handling missing values.

### Identifying Rows with Missing Tuition Fees
1. **Calculate Mean Tuition fees wrt type:**
   - Calculate the mean tuition fees from the 'tuition fees' column with respect to type.

2. **Fill Missing Values:**
   - Replace missing values in the 'tuition fees' column with the calculated mean.

```python
empty_tuitionfee=df.loc[df['tuition_fee'].isna()]
empty_tuitionfee
mean_fee=df.groupby('type')['tuition_fee'].mean()
mean_fee=round(mean_fee)
df['tuition_fee']=df.groupby('type')['tuition_fee'].transform(lambda x:x.fillna(round(x.mean())))
df.isna().sum()
```

## Data Preprocessing: Handling Missing Values

### Calculate Mean Rating and Fill Missing Values:

1. **Calculate Mean Rating:**
   - Calculate the mean rating from the 'rating' column.

2. **Fill Missing Values:**
   - Replace missing values in the 'rating' column with the calculated mean.

```python
mean_fee=round(mean_fee)
rating_mean = df['rating'].mean()
df['rating'].fillna(rating_mean, inplace=True)
df['rating'] = df['rating'].astype(int)
df.isna().sum()
```

## Scatterplot of Applications with regression lines to determine the trend


## 1. Data Visualization Trends

### Scatter Plot with Regression Lines
The scatter plot depicts the relationship between 'tuition_fee' and 'date' for Stem and non-Stem schools. The x-axis represents the date, the y-axis represents the tuition fee, and data points are color-coded by school type. Regression lines for Stem and non-Stem schools reveal potential trends over time.

#### Stem Schools
The black regression line for Stem schools suggests a negative correlation between 'numeric_date' and 'tuition_fee'. As 'numeric_date' increases, 'tuition_fee' tends to be downward.

#### Non-Stem Schools
The red regression line for non-Stem schools shows a different trend. It indicates a flatter relationship between 'numeric_date' and 'tuition_fee', suggesting a less increase compared to Stem schools.

### Conclusion

The scatter plot with regression lines reveals distinct trends in tuition fees over time for Stem and non-Stem schools. Stem schools demonstrate a more pronounced decrease, while non-Stem schools show a more moderate trend. This analysis provides valuable insights into the dynamics of tuition fees in different types of schools.


```python
plt.figure(figsize=(12, 9))
df['date'] = pd.to_datetime(df['date'])

sns.scatterplot(x='date', y='tuition_fee',data=df, hue='type', marker='.', s=100)
df['numeric_date'] = df['date'].apply(mdates.date2num)

sns.regplot(x='numeric_date', y='tuition_fee', data=df[df['type'] == 'Stem'], scatter=False, color='black', line_kws={'linewidth': 1},label='Stem')
sns.regplot(x='numeric_date', y='tuition_fee', data=df[df['type'] == 'non-Stem'], scatter=False, color='red',line_kws={'linewidth': 1},label='non Stem')

plt.legend(title='type', loc='best')
plt.title("Scatter Plot with Regression Lines for Each Type")
plt.xlabel("Date")
plt.ylabel("Tuition Fee")
```
